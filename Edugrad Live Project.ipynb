```python
sns.catplot(x=' is_weekend',y=' shares',data=dat)
plt.title('relation betweeen day of release and shares')
```

From the above graph we can clearly see that articles published on weekdays were shared more compared to articles shared on weekends

```python
def getcount(col):
    result=dat[dat[col]==1][col].value_counts()
    return result

mon=getcount(' weekday_is_monday')
tue=getcount(' weekday_is_tuesday')
wed=getcount(' weekday_is_wednesday')
thu=getcount(' weekday_is_thursday')
fri=getcount(' weekday_is_friday')
sat=getcount(' weekday_is_saturday')
sun=getcount(' weekday_is_sunday')

df=pd.DataFrame([mon,tue,wed,thu,fri,sat,sun])
plt.figure(figsize=(20,9))
df.plot(kind='bar',color='violet')
```

The above graph displays how many articles were published on each day of the week

```python
def getshares(col):
    result=dat[dat[col]==1][' shares'].mean()
    return result

lf=getshares(' data_channel_is_lifestyle')
bus=getshares(' data_channel_is_bus')
socmed=getshares(' data_channel_is_socmed')
tech=getshares(' data_channel_is_tech')
world=getshares(' data_channel_is_world')

df=pd.DataFrame([lf,bus,socmed,tech,world])
df.index=['lifestyle','business','social media','tech','world']
plt.figure(figsize=(20,9))
df.plot(kind='bar',color='yellow')
```

Above graph represents average number of shares of each data channel
